### 컬럼 타입변환
Datetime으로의 컬럼 타입변환 및 기타 object 컬럼들에 대한 타입 변환을 실행한다.

```python
# 일시 -> datetime type 변환
df['일시'] = pd.to_datetime(df['일시'])
```

```python
# object 컬럼들 확인
object_df = df.select_dtypes(include='object')
```

```python
object_df.head()
```

```python
for col in object_df.columns:
    for val in object_df[col].values:
        if '.' in val:
            print(val)
```

```python
# object to int
for col in df.columns:
    if str(df[col].dtype) == "object":
        df[col] = df[col].apply(lambda x : int(x.replace(",", "")))
        
print(df.select_dtypes(include="object").shape[1])
```

### 상수컬럼 여부 및 불필요한 데이터 제거

각 컬럼별로 상수 컬럼의 존재여부를 확인하고, 0으로 채워진 데이터 등의 불필요한 데이터를 제거한다.

```python
df[df['폐기물_거래량'] != 0][['일시', '폐기물_거래량']]  # 폐기물 거래량의 경우 24년 8월 한달 데이터만 포함함
```

```python
[val for val in df.columns if "폐기" in val]  # 폐기물 관련 컬럼 확인
```

```python
# 폐기물 관련된 데이터의 경우 제거
df.drop([val for val in df.columns if "폐기" in val], axis=1, inplace=True)
```

```python
df[df['일시'] >= '2024-12-01']  # 12월의 환경 변수 데이터 확인
```

```python
# 24년 12월의 환경변수의 경우 0값만 있기에, 12월 이전의 데이터만 포함
df_before_20241201 = df[df['일시']<='2024-12-01']
```

```python
df_before_20241201
```

```python
# 신재생 에너지 항목이 아닌 것들은 제외
drop_target = []  # 제거 대상을 담을 리스트
non_rec_sub = ['원자력', '유연탄', '유류', 'LNG', '양수', '기타', '무연탄']  # 비신재생 대상 keywords
for val in non_rec_sub:
    for col in df_before_20241201.columns:  # 기존 데이터들의 컬럼에서
        if val in col:  # 비신재생 대상 키워드가 포함된 데이터는
            drop_target.append(col)  # 제거 대상 리스트에 추가하고 
```

```python
drop_target.pop(0)  # 잘못 담긴 대상은 제외하고 
```

```python
df = df_before_20241201.drop(drop_target, axis=1)  # 원 데이터프레임에서 제외 
```

### 계절적인 요인을 포함하는 변수들 시각화

```python
seasonal_features = [col for col in df.columns if '거래량' in col]
```

```python
fig = go.Figure()
for col in seasonal_features:
    fig.add_trace(go.Scatter(x=df['일시'], y=df[col], mode='lines', name=col))

fig.update_layout(
    width=800, 
    height=400, 
    title="계절적 변수들 시각화"
    )
fig.update_xaxes(title_text="기간")
fig.update_yaxes(title_text="거래량")
fig.show()
```
