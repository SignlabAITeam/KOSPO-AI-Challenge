# 논문 기준 REC 거래 데이터에서 추출한 패턴 데이터


 - REC 거래 규모에 대한 Cyclic Pattern 지수

 - REC 체결 건수에 대한 Cyclic Pattern 지수

 - REC 체결 가격에 대한 Cyclic Pattern 지수

```python
from prophet import Prophet
import pandas as pd

# 데이터 준비
df['현물시장거래일'] = pd.to_datetime(df['현물시장거래일'])

# 분석 대상 변수

# 거래 규모
ts_df = df[['현물시장거래일','육지 체결물량']].rename(columns={'현물시장거래일':'ds','육지 체결물량':'y'})

# 체결 건수
nc_df = df[['현물시장거래일','육지 체결건수']].rename(columns={'현물시장거래일':'ds','육지 체결건수':'y'})

# 체결 가격
ef_df = df[['현물시장거래일','육지 평균가']].rename(columns={'현물시장거래일':'ds','육지 평균가':'y'})
```

```python
ts_df
```

```python
ts_df.info()
```

```python
# 거래 규모
ts_model = Prophet(yearly_seasonality=True, weekly_seasonality=True, daily_seasonality=False)
ts_model.fit(ts_df) 

# 체결 건수
nc_model = Prophet(yearly_seasonality=True, weekly_seasonality=True, daily_seasonality=False)
nc_model.fit(nc_df) 

# 체결 가격
ef_model = Prophet(yearly_seasonality=True, weekly_seasonality=True, daily_seasonality=False)
ef_model.fit(ef_df) 
```

```python
# 미래 데이터 생성
future_ts = ts_model.make_future_dataframe(periods=365)
future_nc = nc_model.make_future_dataframe(periods=365)
future_ef = ef_model.make_future_dataframe(periods=365)

# 예측
forcast_ts = ts_model.predict(future_ts)
forcast_nc = nc_model.predict(future_nc)
forcast_ef = ef_model.predict(future_ef)
```

```python
forcast_ts 
```

```python
# 육지 체결 가격 트렌드 및 주기 성분 
df = df.merge(forcast_ef[['ds','trend','yearly']], left_on='현물시장거래일', right_on='ds', how='left')
df.rename(columns={'trend':'육지_체결_가격_트랜드', 'yearly':'육지_체결_가격_주기성분'}, inplace=True)

# 육지 체결 건수 트렌드 및 주기 성분
df = df.merge(forcast_nc[['ds','trend','yearly']], left_on='현물시장거래일', right_on='ds', how='left')
df.rename(columns={'trend':'육지_체결_건수_트랜드', 'yearly':'육지_체결_건수_주기성분'}, inplace=True)

# 육지 거래 규모 트렌드 및 주기 성분
df = df.merge(forcast_ts[['ds','trend','yearly']], left_on='현물시장거래일', right_on='ds', how='left')
df.rename(columns={'trend':'육지_거래_규모_트랜드', 'yearly':'육지_거래_규모_주기성분'}, inplace=True)
```
