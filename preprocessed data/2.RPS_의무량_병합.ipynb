```python
# df 불러온 뒤,
df['현물시장거래일'] = pd.to_datetime(df['현물시장거래일'])
df['년도'] = df['현물시장거래일'].dt.year
```

```python
RPS_2020 = pd.read_csv(r'C:\Users\PC027\Desktop\Project\KOSPO공모전\REC 예측 관련데이터\RPS 공급의무자별 의무공급량\RPS_2020.csv',encoding='cp949')
RPS_2021 = pd.read_csv(r'C:\Users\PC027\Desktop\Project\KOSPO공모전\REC 예측 관련데이터\RPS 공급의무자별 의무공급량\RPS_2021.csv',encoding='cp949')
RPS_2022 = pd.read_csv(r'C:\Users\PC027\Desktop\Project\KOSPO공모전\REC 예측 관련데이터\RPS 공급의무자별 의무공급량\RPS_2022.csv',encoding='cp949')
RPS_2023 = pd.read_csv(r'C:\Users\PC027\Desktop\Project\KOSPO공모전\REC 예측 관련데이터\RPS 공급의무자별 의무공급량\RPS_2023.csv',encoding='cp949')
RPS_2024 = pd.read_csv(r'C:\Users\PC027\Desktop\Project\KOSPO공모전\REC 예측 관련데이터\RPS 공급의무자별 의무공급량\RPS_2024.csv',encoding='cp949')
```

```python
RPS_2021
```

```python
# RPS 데이터 합치기
RPS_total = pd.concat(
    [RPS_2020, RPS_2021, RPS_2022, RPS_2023, RPS_2024],
    ignore_index=True
)
```

```python
# 각 RPS 데이터에서 기업명 추출
companies_2020 = set(RPS_2020["기업명"].unique())
companies_2021 = set(RPS_2021["기업명"].unique())
companies_2022 = set(RPS_2022["기업명"].unique())
companies_2023 = set(RPS_2023["기업명"].unique())
companies_2024 = set(RPS_2024["기업명"].unique())

# 5개 년도에 모두 존재하는 기업명
common_companies = (
    companies_2020 & 
    companies_2021 & 
    companies_2022 & 
    companies_2023 & 
    companies_2024
)
```

```python
RPS_total['년도'].unique()
```

```python
RPS_total.columns
```

```python
common_companies 
```

```python
# 위에서 구한 common_companies 만 필터
RPS_filtered = RPS_total[RPS_total["기업명"].isin(common_companies)]
```

```python
# 원하는 컬럼을 피벗하여 기업명별 열 생성
pivot_rec = RPS_filtered.pivot(
    index='년도', 
    columns='기업명', 
    values='의무공급량(REC)'
)
```

```python
pivot_rec
```

```python
pivot_rec.columns = [str(col) + '_RPS' for col in pivot_rec.columns]
```

```python
pivot_rec
```

```python
# df의 현물시장거래일로부터 년도 추출
df['현물시장거래일'] = pd.to_datetime(df['현물시장거래일'])
df['년도'] = df['현물시장거래일'].dt.year

# pivot_mwh의 인덱스 -> 컬럼
pivot_mwh = pivot_rec.reset_index()

df_merged = pd.merge(
    df,
    pivot_mwh,
    how='left',  # 필요에 따라 'inner' 등 변경
    on='년도'
)
```

```python
df_merged
```

```python
df_merged.columns
```

```python
df_merged['년도'].unique()
```

```python
# 2017, 2018, 2019 데이터를 제거
df_filtered = df_merged[~df_merged['년도'].isin([2017, 2018, 2019])]
```
