```python
# 데이터프레임의 0번 인덱스를 컬럼으로 설정하는 함수
def set_first_row_as_columns(df):
    df.columns = df.iloc[0]  # 0번 인덱스를 컬럼으로 설정
    df = df[1:].reset_index(drop=True)  # 0번 인덱스 데이터 제거 후 인덱스 리셋
    return df

# 데이터프레임 리스트
dataframes = [smp_2020,smp_2021,smp_2022, smp_2023, smp_2024]

# 각 데이터프레임에 함수 적용
smp_2020, smp_2021, smp_2022, smp_2023, smp_2024 = [set_first_row_as_columns(df) for df in dataframes]
```

```python
smp_2020.head()
```

```python
smp_combined = pd.concat([smp_2020, smp_2021, smp_2022, smp_2023, smp_2024], ignore_index=True)
smp_combined
```

```python
# 최소, 최대, 평균 컬럼 이름 변경
smp_combined.rename(columns={
    '최소': 'SMP_최소',
    '최대': 'SMP_최대',
    '평균': 'SMP_평균',
    '구분' : '현물시장거래일'
}, inplace=True)

print(smp_combined.columns)
```

```python
# int64 형식의 날짜를 문자열로 변환 후, pd.to_datetime 사용
smp_combined['현물시장거래일'] = pd.to_datetime(smp_combined['현물시장거래일'].astype(str), format='%Y%m%d')
```

```python
df['현물시장거래일'] = pd.to_datetime(df['현물시장거래일'])
```

```python
# 필요한 컬럼만 선택 (현물시장거래일, SMP_최소, SMP_최대, SMP_평균)
smp_combined_subset = smp_combined[['현물시장거래일', 'SMP_최소', 'SMP_최대', 'SMP_평균']]

# df와 smp_combined 병합 (현물시장거래일 기준)
df = pd.merge(df, smp_combined_subset, on='현물시장거래일', how='left')

df.head()
```

```python
# NaN 값이 하나라도 있는지 확인
has_nan = df.isnull().values.any()

print(f"NaN 값이 존재합니다: {has_nan}")
```

```python
# 각 열의 NaN 값 개수 확인
nan_counts = df.isnull().sum()

print("각 열의 NaN 값 개수:")
print(nan_counts)
```

```python
# NaN 값이 있는 열만 필터링
nan_counts = df.isnull().sum()
nan_counts = nan_counts[nan_counts > 0]  # NaN 값이 1개 이상 있는 열만 선택

print("NaN 값이 있는 열과 개수:")
print(nan_counts)
```
