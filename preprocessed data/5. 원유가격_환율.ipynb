# 원유가격 코드 작성


```python
co_1 = pd.read_csv(r'C:\Users\PC027\Desktop\Project\KOSPO공모전\REC 예측 관련데이터\원유가격\일일국제원유가격(200102~211231).csv')
```

```python
co_1 
```

```python
co_1 = co_1.drop('유종',axis=1)
```

```python
co_1.tail(10)
```

```python
co_1 = co_1.iloc[:518]
```

```python
co_1 = co_1.rename(columns={'﻿일':'일'})
```

```python
# 초기 연도 설정
current_year = 2020
years = []

# 날짜 순서대로 연도 할당
for idx, row in co_1.iterrows():
    if '01월' in row['일'] and idx > 0 and '12월' in co_1.loc[idx - 1, '일']:
        current_year += 1  # 연도 증가
    years.append(current_year)

# 연도를 '일'에 추가하여 날짜 변환 (월과 일을 정확히 포맷팅)
co_1['일'] = [f"{year}-{date.split('월')[0].zfill(2)}-{date.split('월')[1].replace('일', '').strip().zfill(2)}"
              for year, date in zip(years, co_1['일'])]
co_1['일'] = pd.to_datetime(co_1['일'], format='%Y-%m-%d')
```

```python
co_1
```

```python
co_2 = pd.read_csv(r'C:\Users\PC027\Desktop\Project\KOSPO공모전\REC 예측 관련데이터\원유가격\일일국제원유가격(220103~241224).csv')
```

```python
co_2.tail(10) 
```

```python
co_2 = oc_2.iloc[:770]
```

```python
co_2 = co_2.rename(columns={'﻿일':'일'})
```

```python
# 초기 연도 설정
current_year = 2022
years = []

# 날짜 순서대로 연도 할당
for idx, row in co_2.iterrows():
    # 12월에서 다음 데이터가 1월로 넘어가면 연도 증가
    if '01월' in row['일'] and idx > 0 and '12월' in co_2.loc[idx - 1, '일']:
        current_year += 1  # 연도 증가
    years.append(current_year)

# 연도를 '일'에 추가하여 날짜 변환 (월과 일을 정확히 포맷팅)
co_2['일'] = [f"{year}-{date.split('월')[0].zfill(2)}-{date.split('월')[1].replace('일', '').strip().zfill(2)}"
              for year, date in zip(years, co_2['일'])]

# '일'을 datetime 형식으로 변환
co_2['일'] = pd.to_datetime(co_2['일'], format='%Y-%m-%d')

# 결과 출력
print(co_2)
```

```python
co_2 = co_2.drop('유종',axis=1)
```

```python
co_2
```

```python
combined_oc = pd.concat([co_1,co_2])
combined_oc = combined_oc.reset_index()
```

```python
combined_oc
```

```python
combined_oc = combined_oc.drop(columns=['index'])
```

```python
combined_oc
```

```python
combined_oc.columns
```

## 데이터에 - 가 들어가면 0으로 치환해서 사용

```python
combined_oc['Dubai'].value_counts()
```

```python
# 네 개의 컬럼에서 '-' 값을 0으로 변환
columns_to_replace = ['Dubai', 'Brent', 'WTI', 'Oman']

# 반복문을 통해 각 컬럼의 '-' 값을 0으로 변경
for col in columns_to_replace:
    combined_oc[col] = combined_oc[col].replace('-', 0).astype(float)

# 결과 출력
print(combined_oc)
```

```python
combined_oc['Dubai'].value_counts()
```

```python
df['현물시장거래일'].info()
```

```python
df['현물시장거래일'] = pd.to_datetime(df['현물시장거래일'])
```

```python
# '일' 열 제외한 co_combined의 나머지 열 선택
co_combined_except_일 = combined_oc.drop(columns=['일'])

# 병합 수행 ('현물시장거래일'과 '일'을 기준으로 병합)
merged_df = pd.merge(
    df,
    co_combined_except_일,
    left_on='현물시장거래일',
    right_on=combined_oc['일'],
    how='left'  # 필요에 따라 'inner', 'right', 'outer'로 변경 가능
)

# 결과 출력
print(merged_df)
```
