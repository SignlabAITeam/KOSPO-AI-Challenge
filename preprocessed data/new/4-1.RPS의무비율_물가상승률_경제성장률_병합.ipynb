# RPS 의무비율 
## 출처 - 한국에너지공단 신재생에너지센터
 - 2017 / 4.0%

 - 2019 / 6.5%
 - 2020 / 7.0%

 - 2023 / 13%

```python
test= df
```

```python
import numpy as np

# 조건 정의
conditions = [
    test['년도'] == 2017,
    test['년도'] == 2018,
    test['년도'] == 2019,
    test['년도'] == 2020,
    test['년도'] == 2021,
    test['년도'] == 2022,
    test['년도'] == 2023,
    test['년도'] == 2024
]

# 각 조건에 할당할 값 정의 
values = [4.0,5.0,6.5,7.0, 9.0, 12.5, 13, 13.5]

# 의무 RPS 비율 컬럼 생성 및 값 할당
test['의무RPS비율'] = np.select(conditions, values, default=np.nan)

test.head()
```
