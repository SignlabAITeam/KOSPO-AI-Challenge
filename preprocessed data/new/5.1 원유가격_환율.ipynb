```python
co['기간'] = pd.to_datetime(co['기간'], format = '%y년%m월%d일')

# 결과 확인
print(co.head())
print(co.info())
```

```python
# 네 개의 컬럼에서 '-' 값을 0으로 변환
columns_to_replace = ['Dubai', 'Brent', 'WTI']

# 반복문을 통해 각 컬럼의 '-' 값을 0으로 변경
for col in columns_to_replace:
    co[col] = co[col].replace('-', 0).astype(float)

# 결과 출력
print(co)
```

```python
df['현물시장거래일'] = pd.to_datetime(df['현물시장거래일'])
```

```python
co.head()
```

```python
# '일' 열 제외한 co_combined의 나머지 열 선택
co_combined_except_일 = co.drop(columns=['기간'])

# 병합 수행 ('현물시장거래일'과 '일'을 기준으로 병합)
merged_df = pd.merge(
    df,
    co_combined_except_일,
    left_on='현물시장거래일',
    right_on=co['기간'],
    how='left'  # 필요에 따라 'inner', 'right', 'outer'로 변경 가능
)

# 결과 출력
print(merged_df)
```
