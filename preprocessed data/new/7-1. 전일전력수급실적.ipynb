```python
import os
import win32com.client

# 경로 설정
path = r'C:\Users\PC027\Desktop\Project\KOSPO공모전\REC 예측 관련데이터\전일전력수급실적'
output_path = path  # 변환된 CSV 파일 저장 경로 (필요 시 변경)

# 디렉토리의 모든 파일 리스트 가져오기
file_list = os.listdir(path)

# .xls 파일만 필터링
xls_files = [file for file in file_list if file.endswith('.xls')]

# Excel Application 객체 생성
excel = win32com.client.Dispatch("Excel.Application")
excel.DisplayAlerts = False  # 파일 저장 시 알림 비활성화

try:
    for xls_file in xls_files:
        xls_path = os.path.join(path, xls_file)  # 원본 파일 경로
        csv_file = os.path.splitext(xls_file)[0] + '.csv'  # 확장자를 .csv로 변경
        csv_path = os.path.join(output_path, csv_file)  # 저장할 파일 경로

        try:
            # Excel 파일 열기
            workbook = excel.Workbooks.Open(xls_path)
            
            # CSV로 저장
            workbook.SaveAs(csv_path, FileFormat=6)  # FileFormat=6은 CSV 형식
            workbook.Close()  # 파일 닫기

            print(f"Converted: {xls_file} -> {csv_file}")
        except Exception as e:
            print(f"Failed to convert {xls_file}: {e}")
finally:
    excel.Quit()  # Excel 프로세스 종료
```

```python
import os
import pandas as pd
from datetime import datetime, timedelta

# ========== 1) 경로 & 파일 목록 설정 ==========
path = r'C:\Users\PC027\Desktop\Project\KOSPO공모전\REC 예측 관련데이터\전일전력수급실적'
file_list = [f for f in os.listdir(path) if f.endswith('.csv')]

# ========== 2) 전처리 함수 ==========
def process_dataframe(df: pd.DataFrame, columns_to_drop: list) -> pd.DataFrame:
    """
    주어진 데이터프레임에 대해 특정 열 제거, 0번 행을 컬럼으로 설정, 초기화 과정 수행
    Args:
        df (pd.DataFrame): 처리할 데이터프레임
        columns_to_drop (list): 제거할 열 이름 목록
    Returns:
        pd.DataFrame: 처리된 데이터프레임
    """
    # 지정된 열 제거
    df = df.drop(columns=columns_to_drop, axis=1)
    
    # 0번 행을 새로운 컬럼명으로 설정
    df.columns = df.iloc[0]
    
    # 0번 행 제거
    df = df.drop(index=0)
    
    # 인덱스 초기화
    df = df.reset_index(drop=True)
    
    # 추가로 1번 행 제거(예시 코드에 나와 있으므로)
    df = df.iloc[1:].reset_index(drop=True)
    return df

# ========== 3) 결측 날짜(NaN) 채우기 함수 ==========
def fill_missing_dates(df: pd.DataFrame, date_col: str = '일시') -> pd.DataFrame:
    """
    '일시' 열의 NaN을 바로 윗행(직전 행)의 날짜 - 1일 로 채운다.
    시간은 의미 없으므로 '00:00'으로 통일.

    Args:
        df (pd.DataFrame): 처리할 데이터프레임
        date_col (str): 날짜가 들어있는 열 이름 (기본값: '일시')
    Returns:
        pd.DataFrame: '일시' 열의 결측이 채워진 데이터프레임
    """

    # 먼저 '일시' 컬럼을 datetime 파싱 시도
    # (CSV가 "YYYY.MM.DD. HH:MM" 형태라면 아래와 같이 포맷 지정. 아니라면 포맷 수정 필요.)
    def try_parse_date(x):
        try:
            return datetime.strptime(x, '%Y.%m.%d. %H:%M')
        except:
            return pd.NaT  # 파싱 실패 시 NaT로
    
    # 문자열로 변환 (혹시 float 등 다른 타입일 수 있으니)
    df[date_col] = df[date_col].astype(str)
    
    # datetime 형으로 변환 시도
    parsed_dates = df[date_col].apply(try_parse_date)
    
    # NaT(결측) 존재 시, 바로 이전 행 기준 하루 전날로 채움
    for i in range(len(parsed_dates)):
        if pd.isna(parsed_dates.iloc[i]):
            if i == 0:
                # 첫 행부터 NaN이라면 임의값(1900-01-01 00:00)으로 채움
                parsed_dates.iloc[i] = datetime(1900, 1, 1, 0, 0)
            else:
                # 직전 날짜에 하루 빼서 할당
                parsed_dates.iloc[i] = parsed_dates.iloc[i - 1] - timedelta(days=1)

    # 최종 날짜 컬럼을 다시 문자열 형태(예: YYYY.MM.DD. 00:00)로 저장
    df[date_col] = parsed_dates.apply(lambda d: d.strftime('%Y.%m.%d. 00:00'))
    
    return df

# ========== 4) 모든 CSV 처리하여 하나로 합치기 ==========
all_data = []
columns_to_drop = ['Unnamed: 4', 'Unnamed: 5', 'Unnamed: 6']  # 예시

for file_name in file_list:
    csv_path = os.path.join(path, file_name)
    
    # CSV 로드
    tmp_df = pd.read_csv(csv_path, encoding='cp949')
    
    # (1) 전처리
    tmp_df = process_dataframe(tmp_df, columns_to_drop)
    
    # (2) 결측 날짜 채우기
    tmp_df = fill_missing_dates(tmp_df, date_col='일시')
    
    # 리스트에 추가
    all_data.append(tmp_df)

# ========== 5) 하나의 데이터프레임으로 병합 ==========
final_df = pd.concat(all_data, ignore_index=True)

# ========== 6) 결과 확인 ==========
print(final_df.head())
print("최종 행 개수:", len(final_df))
```

```python
import pandas as pd

# 1. final_df의 '일시' 열을 날짜 형식으로 변환
final_df['일시_변환'] = pd.to_datetime(final_df['일시'], format='%Y.%m.%d. %H:%M').dt.date

# 2. df의 '현물시장거래일'을 날짜 형식으로 변환
df['현물시장거래일_변환'] = pd.to_datetime(df['현물시장거래일']).dt.date

# 3. 병합 수행
merged_df = pd.merge(final_df, df, left_on='일시_변환', right_on='현물시장거래일_변환', how='inner')

# 4. df 기준으로 고유한 행만 유지
merged_df = merged_df.drop_duplicates(subset=['현물시장거래일_변환'])

# 결과 출력
merged_df
```
