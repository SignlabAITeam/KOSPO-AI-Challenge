```python
# 변환 함수 정의
def convert_period(period):
    # 연도와 월을 분리
    year, month = period.split('-')  # 수정: 연도와 월 순서로 분리
    # 년도가 50 이상이면 1900년대, 그 외에는 2000년대
    if int(year) >= 50:  # 연도를 판단할 기준 설정 (50 이상은 1900년대)
        full_year = f"19{year}"
    else:
        full_year = f"20{year}"
    # 월 이름을 숫자로 변환
    month_dict = {
        'Jan': '01', 'Feb': '02', 'Mar': '03', 'Apr': '04', 'May': '05', 'Jun': '06',
        'Jul': '07', 'Aug': '08', 'Sep': '09', 'Oct': '10', 'Nov': '11', 'Dec': '12'
    }
    full_month = month_dict[month]
    return f"{full_year}-{full_month}"

# '기간' 컬럼 변환 적용
et['변환된 기간'] = et['기간'].apply(convert_period)
```
